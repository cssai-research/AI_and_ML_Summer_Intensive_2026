### map() Function

map applies a function to all the elemements of a list / collection

```python
def square(num):
  '''To square the number'''
  return num**2
```

```python
my_nums = range(0,20)
```

```python
square_of_numbers = list(map(square,my_nums))
square_of_numbers
```

```python
for item in map(square, my_nums):
  print(item)
```

### lambda() Function

Lambda functions are function that can be called with a variable number of parameters. They are written in a more concise way  

```python
# We will convert the following function into a lambda function.
def square(num):
  return num**2
```

```python
square(9)
```

```python
square_lambda = lambda n : n**2
```

```python
square_lambda(9)
```

```python
num = [1,2,3,4,5,6]
```

```python
list(map(lambda num : num**2, num))
```

```python
# Lambda expression for grabbing the first character of a string
string_firstchar = lambda s: s[0]
```

```python
string_firstchar('Hello')
```

```python
# Lambda expression for reversing a string
string_rev = lambda s: s[::-1]
```

```python
string_rev('New York')
```

```python
#Pass in multiple arguments into a lambda expression.
add = lambda x,y : x + y
add(50,60)
```

### *args and **kwargs

```python
def myfunc(a,b):
    return sum((a,b))

myfunc(40,60)
```

```python
def myfunc(a,b,c):
    return sum((a,b,c))

myfunc(40,60,100)
```

In this example, a and b are positional arguments; that is, 40 is assigned to **a** because it is the first argument, and 60 to **b**.

What if we want to work with more than two numbers?

    * One way would be to assign a lot of parameters and give each
      one a default value.
    * Use  *args

```python
def myfunc(a=0,b=0,c=0,d=0,e=0):
    return sum((a,b,c,d,e))

myfunc(40,60,20)
```

```python
def myfunc_args(*args):
    return sum(args)
```

```python
myfunc_args(40,60,20)
```

```python
myfunc_args(40,60,20,50)
```

```python
myfunc_args(40,60,20,30,50, 10)
```

 #### **kwargs

kwargs allows a function to accept an arbitrary number of keyword arguments

```python
def myfunc_kwargs(**kwargs):
  if 'animal' in kwargs:
    print("My favorate animal is {}".format(kwargs['animal']))
  elif 'car' in kwargs:
    print("My favorate car is {}".format(kwargs['car']))
  else:
    print("Not found")
```

```python
myfunc_kwargs(car='Tesla')
```

```python
def myfunc_args_kwargs(*args, **kwargs):
  print(args)
  print(kwargs)
  print("I want {} {}".format(args[0],kwargs['car']))
```

```python
myfunc_args_kwargs(10,50, car='Tesla', food="salad", animal='cat')
```
